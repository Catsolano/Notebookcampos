# Carga inalambrica

La carga inalambrica, tambien conocida como tecnología Qi,  es aquella en la cual se busca la transferencia de energía mediante un campo electromagnético. El acoplamiento inductivo resonante es el cual describe la transmisión de energía eléctrica inalámbrica cercana entre dos bobinas acopladas magnéticamente y que resuenan en la misma frecuencia. 

La inducción magnética es la generación de un campo eléctrico debido a un campo magnético que varía con el tiempo, tal y como demostró Faraday. El voltaje inducido en un circuito cerrado es directamente proporcional a la rapidez de la variación el flujo magnético que atraviesa una superficie cualquiera, es decir: 

 \begin{equation}
(1)  \oint E* dl = - \frac{d}{dt} \oint B* dA 
\end{equation}


La tecnología Qi está constituida por un sistema de dos bobinas, una Transmisora (Tx) y una receptora(Rx), con principio de acoplamiento resonante.


### Factor de acoplamiento k:
Es fundamental a la hora de realizar un estudio de acoplamiento magnético ya que este factor brinda un valor entre 0 y 1. Cuando se presenta el valor de 0 se puede concluir que las bobinas están aisladas magnéticamente, por el contrario, cuando el valor es 1 estan perfectamente acopladas.

Una variable importante a la hora de calcular el factor de acoplamiento k es la distancia entre ambas bobinas, se dice que la 
distancia entre ambas no puede superar el diametro de la bobina emisora o sino será ineficiente. 

\begin{equation}
(2) k = \frac{M}{\sqrt{L_1 L_2}}
\end{equation}
 


### Inductancia mutua: 
Es la capacidad de un inductor de producir caida de tensión en un inductor cercano, debido al flujo magnético que enlaza el primer inductor y el número de vueltas del conductor del segundo inductor donde su unidad es el Henry (H).


 \begin{equation}
(3) V_x = M_{xs} \sqrt{di}{dt}  
\end{equation}

Tomando a x como el indicativo de la tensión de la bobina que estamos hablando, sea 1 u 2. El subindice x indica que la inductancia Mx (bobina 1 u 2) se relaciona con la tensión inducida por la s, con corriente en la bobina x.

Ahora, se tiene que la formula para hallar la inductancia mutua, despejada de la ecuación (1) es:

 \begin{equation}
(4) M = k \sqrt{L_1 L_2} = \frac{k}{w} \sqrt{Q_1 Q_2 R_1 R_2}   
\end{equation}


La fem inducida en una bobina se describe mediante la ley de Faraday y su dirección SIEMPRE es opuesta al cambio del campo magnetico producido en ella por la bobina acoplada (Ley de Lenz).

La ley de Faraday es:

\begin{equation}
(5)  \phi _m  = \iint \overrightarrow{B} \overrightarrow{ds}  
\end{equation}

La ecuación anterior se puede usar para una sola espira. Ahora si lo queremos usar con varias bobinas debemos remitirnos a la siguiente ecuación:

\begin{equation}
(6) e = fem = -N L \iint \overrightarrow{B} \overrightarrow{ds}  = -N \frac{ \Delta (B*A)} {\Delta t} 
\end{equation}

Fem es el voltaje generado o inducido. Se tiene que N son el número de bobinas/espiras; y L es la longitud de la bobina. 

### Energía en un circuito acoplado.

Se tiene que la energía almacenada por un inductor está dada por:

 \begin{equation}
(7)  \omega = {\frac{1}{2}} L_{1^{2}}   
\end{equation}

Ahora interesa determinar la energía almacenada en bobinas magnéticamente acopladas. Dada por la expresión:


 \begin{equation}
(8) \omega = \frac{1}{2} L_1 I_{1^{2}}+ \frac{1}{2} L_1 I_{2^{2}}+ M I_1 I_2   
\end{equation}

Tomando el siguiente circuito se va a analizar paso por paso lo conveniente a seguir para concluir la ecuación final.

El primer paso a seguir es analizar la calidad del inductor con la siguiente formula:

 \begin{equation}  
 (9) Q = \frac{fo}{f2-f1} = \frac{fo}{\Delta f}    
\end{equation}

Donde fo es la frecuencia de resonancia del circuito, esta formula es usada en componentes no ideales.
Si Q tiene un valor alto, hay una perdida de energía.

Se debe estudiar el factor de calidad de cada bobina, tanto la que recibe como la emisora.

 \begin{equation}
(10)  Q_n  = \frac{ \omega L_n}{R_n}                       
\end{equation}

n se refiere a la generalización de la bobina a la que nos referimos, ya sea la 1 emisora o la 2 receptora.

Se debe tener en cuenta el concepto de inductancia mutua, expresada por la ecuación (4).

Vamos a analizar el circuito realizando mallas tanto en el lado primario como en el secundario.

 \begin{equation}
(11)  V_in  = R_1 I_1 + \frac{I_1}{ j \omega C_1}+ j\omega L_1 I_1 - j \omega M I_2                       
\end{equation}

  \begin{equation}
(12)  j \omega M I_1 = R_2 I_2 + \frac{I_2}{ j \omega C_2}+ j\omega L_2 I_2 - R_L I_2                       
\end{equation}

Se eligen los componentes C1, L1, C2, L2 de tal forma que resuenen entre sí.

 \begin{equation}
(13)  \omega^{2} = \frac{1}{ L_1 C_1} =   \frac{1}{ L_2 C_2}                    
\end{equation}

Los términos correspondientes a L1, L2 se anulan con los términos correspondientes de C1, C2:
  \begin{equation}
(14)  V_in= R_1 I_1+ \frac{\omega^{2} M^{2}}{R_2+R_L} I_1 = (R_1+R_V)I_1                     
\end{equation}

Teniendo a Rv como la resistencia vista desde el primario hacia el secundario.

 \begin{equation}
(15)  R_v= \frac{\omega^{2} M^{2}}{R_2+R_L}                    
\end{equation}

Recordando las definiciones dadas para el factor de calidad Q y la inductancia mutua, se puede concluir a w^2 M^2 como:

 \begin{equation}
(16)  \omega^{2} M^{2}= k^{2} R_1 R_2 Q_1 Q_2                    
\end{equation}

Ahora teniendo en cuenta la ecuación (11) expresamos a Rv así:

 \begin{equation}
(17)  R_v = \frac{k^{2} R_1 R_2 Q_1 Q_2}{ R_2+R_L} =   k^{2} R_1 Q_1 Q_2L                    
\end{equation}

Finalmente, tomando los dos bloques como divisores resistivos se puede expresar la eficiencia del sistema como la multiplicación de las eficiencias de cada bloque. Si se calcula la eficiencia de cada divisor resistivo se obtiene:

  \begin{equation}
(18)  n_1 = \frac{R_v}{ R_1+ R_v} = \frac{ k^{2} Q_1 Q_2L}{1 + k^{2} Q_1 Q_2L}
\end{equation}

  \begin{equation}
(19)  n_2 = \frac{R_L}{ R_2+ R_L} = \frac{ Q_2L}{Q_L}
\end{equation}

Dado que n = n1n2 entonces

  \begin{equation}
(20)  n= \frac{R_v}{ R_1+ R_v} * \frac{ R_L}{R_2 + R_L}
\end{equation}

Expresando la ecuacion (16) en factores de calidad:

  \begin{equation}
(21)  n = \frac{k^{2} Q_2 Q_2L }{1+ k^{2} Q_1 Q_2L} * \frac{Q_2L}{Q_L}
\end{equation}

La ecuación 17 representa la eficiencia intrínseca del acople inductivo para bobinas con resistencias no nulas. Luego, buscamos los máximos respecto a Rl. Así se obtiene una única solución donde la eficiencia es máxima.

  \begin{equation}
(22)  Q_{Lnmax} = \frac{Q_2}{\sqrt{1+ k^{2} Q_1 Q_2}} 
\end{equation}

   \begin{equation}
(23)  R_{Lnmax} = R_2 \sqrt{1+ k^{2} Q_1 Q_2}
\end{equation}

Uniendo las ecuaciones (18) y (19) obtenemos el punto de eficiencia máxima.

  \begin{equation}
(24)  n_{max} = \frac{k^{2} Q_1 Q_2}{1+ \sqrt{1+ k^{2} Q_1 Q_2}} 
\end{equation}

## Máxima transferencia de potencia.

Una vez estudiados los temas previos surge la pregunta si un sistema como este si es capaz de suministrar la potencia necesaria para que el sistema funcione eficientemente. Así que se hará una deducción de las ecuaciones que nos permitirán identificar en cual punto y que variables afectan para lograr la potencia necesaria.

Vemos la potencia de salida como:

  \begin{equation}
(25)  P_{out} = n P_{in} = n V_{in} I_{in} = V^{2}_{in} * \frac{n}{R_1+R_v} =  V^{2}_{in}* \frac{R_v}{(R_1+R_v)^{2}}* \frac{R_L}{R_2+R_L}
\end{equation}

Ahora se sustituye el valor de Rv

 \begin{equation}
(26)  P_{out} = \frac{V^{2}_{in} k^{2} Q_1 Q_2 R_2}{R_1} * \frac{R_L}{(R_L+R_2*(1+ k^{2} Q_1 Q_2))^{2}}
\end{equation}

Luego podemos concluir la potencia de salida como:

\begin{equation}
(27)  P_{out} = A* \frac{R_L}{(R_L+R_a)^{2}}
\end{equation}

Donde
\begin{equation}
 (28) R_a= R_2(1+k^{2} Q_1 Q_2)
\end{equation}

\begin{equation}
 (29)A = \frac{V^{2}_{in} k^{2} Q_1 Q_2 R_2}{R_1} 
\end{equation}

Se halla el punto de máxima transferencia de potencia donde la derivada de Pout respecto Rl se hace 0.

\begin{equation}
(30)  \frac{dP_{out}}{dR_L} = A* \frac{R_a - R_L}{(R_L+R_a)^{3}}
\end{equation}

Así que se anula la derivada en Rl=Ra. Entonces la resistencia vista de carga que maximiza la potencia entregada a la carga es:

\begin{equation}
(31)  R_{Pmax} = R_2 *(1+ k^{2} Q_1 Q_2)
\end{equation}
